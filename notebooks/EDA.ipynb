```python
import seaborn as sns  
import matplotlib.pyplot as plt  

plt.figure(figsize=(6,4))
    #ploting numerical data
for col in numerical_column:
    sns.histplot(data=df,x=col,kde=True,)
    print('\n')
    plt.show()
```

```python
plt.figure(figsize=(20, 20))  # Width: 12, Height: 8 (you can adjust these values)
sns.heatmap(df[numerical_column].corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")  # Optional: Add a title
plt.show()
```
