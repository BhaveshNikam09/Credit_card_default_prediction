# Data Info
1.ID: ID of each client

2.LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit

3.SEX: Gender (1=male, 2=female)

4.EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)

5.MARRIAGE: Marital status (1=married, 2=single, 3=others)

6.AGE: Age in years

7.PAY_1: Repayment status in September, 2005 

8.PAY_2: Repayment status in August, 2005 (scale same as above)

9.PAY_3: Repayment status in July, 2005 (scale same as above)

10.PAY_4: Repayment status in June, 2005 (scale same as above)

11.PAY_5: Repayment status in May, 2005 (scale same as above)

12.PAY_6: Repayment status in April, 2005 (scale same as above)

13.BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)

14.BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)

15.BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)

16.BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)

17.BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)

18.BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)

19.PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)

20.PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)

21.PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)

22.PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)

23.PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)

24.PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)

25.default.payment.next.month: Default payment (1=yes, 0=no)

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Convert PAY_X columns to numeric (since they are objects)
pay_columns = ['PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
df[pay_columns] = df[pay_columns].apply(pd.to_numeric, errors='coerce')

```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['LIMIT_BAL'], bins=30, kde=True, color='blue')
plt.title('Distribution of Credit Limit')
plt.xlabel('Credit Limit (LIMIT_BAL)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=df['default.payment.next.month'], palette=['green', 'red'])
plt.title("Default Payment Next Month")
plt.xlabel("Default (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["AGE"], color='orange')
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='SEX', y='default.payment.next.month', data=df, palette=['purple', 'pink'])
plt.xticks([0,1], ['Male', 'Female'])
plt.title('Default Rate by Gender')
plt.ylabel('Default Probability')
plt.xlabel('Gender')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='default.payment.next.month', y='LIMIT_BAL', data=df, palette=['blue', 'red'])
plt.title("Credit Limit Distribution by Default Status")
plt.xlabel("Default (0 = No, 1 = Yes)")
plt.ylabel("Credit Limit")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='default.payment.next.month', y='BILL_AMT1', data=df, palette=['green', 'red'])
plt.title("Bill Amount 1 Distribution by Default Status")
plt.xlabel("Default Pay")
plt.ylabel("Bill Amount ")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['BILL_AMT1'], y=df['PAY_AMT1'], hue=df['default.payment.next.month'], palette=['blue', 'red'])
plt.title("Bill Amount vs Payment Amount")
plt.xlabel("Bill Amount (BILL_AMT1)")
plt.ylabel("Payment Amount (PAY_AMT1)")
plt.show()
```

```python
sns.pairplot(df[['LIMIT_BAL', 'BILL_AMT1', 'PAY_AMT1', 'AGE', 'default.payment.next.month']], hue="default.payment.next.month", palette=['green', 'red'])
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
edu_counts = df.groupby(['EDUCATION', 'default.payment.next.month']).size().unstack()
edu_counts.plot(kind='bar', stacked=True, colormap="coolwarm")
plt.title("Default Rate by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.legend(["No Default", "Default"])
plt.show()
```

```python
plt.figure(figsize=(30,5))
sns.boxplot(x='PAY_1', y='LIMIT_BAL', data=df, palette='coolwarm')
plt.title("Credit Limit vs Repayment Status (PAY_1)")
plt.xlabel("Repayment Status (PAY_1)")
plt.ylabel("Credit Limit")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df[['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']].mean().plot(marker='o', linestyle='-')
plt.title("Average Bill Amount Over Time")
plt.xlabel("Months (1 to 6)")
plt.ylabel("Average Bill Amount")
plt.show()
```

```python


# Loop through numerical columns and plot histograms
for col in numerical_column:
    plt.figure(figsize=(8, 4))
    sns.histplot(data=df, x=col, hue='default.payment.next.month', kde=True, palette=['green', 'red'])
    plt.title(f"Distribution of {col} by Default Status")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.legend(["No Default", "Default"])
    plt.show()
```
